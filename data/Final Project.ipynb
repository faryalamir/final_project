```python
file1 = pd.read_csv('airlines.csv')
file2 = pd.read_csv('airports.csv')
file3 = pd.read_csv('flights.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the number of flights with departure delays
departure_delays_count = len(file3[file3['DEPARTURE_DELAY'] > 0])

# Calculate the number of flights with arrival delays
arrival_delays_count = len(file3[file3['ARRIVAL_DELAY'] > 0])

# Determine the primary reason for delays
primary_delay_reason = 'Departure' if departure_delays_count > arrival_delays_count else 'Arrival' if arrival_delays_count > departure_delays_count else 'None'

# Create a bar chart with different shades of blue
delay_counts = [departure_delays_count, arrival_delays_count]
labels = ['Departure Delays', 'Arrival Delays']

colors = ['b', 'dodgerblue']  # Different shades of blue

plt.bar(labels, delay_counts, color=colors)
plt.xlabel('Delay Type')
plt.ylabel('Number of Flights')
plt.title('Comparison of Departure Delays and Arrival Delays')
plt.show()

# Print the results
print(f"Number of Flights with Departure Delays: {departure_delays_count}")
print(f"Number of Flights with Arrival Delays: {arrival_delays_count}")
print(f"The primary reason for delays is: {primary_delay_reason}")
```

#### How many unique airlines are there in the dataset, and what are their names?



```python
import pandas as pd
unique_airlines = file1['AIRLINE'].unique()
num_unique_airlines = len(unique_airlines)

print(f"There are {num_unique_airlines} unique airlines in the dataset.")
print("Airline Names:")
for airline in unique_airlines:
    print(airline)
```

#### What is the average departure delay for each airline in the dataset?



```python
average_delay_by_airline = file3.groupby('AIRLINE')['DEPARTURE_DELAY'].mean().reset_index()

print("Average Departure Delay by Airline:")
print(average_delay_by_airline)
```

#### Which origin airport has the highest number of departures?



```python
most_departures = file3['ORIGIN_AIRPORT'].value_counts().idxmax()

print(f"The origin airport with the most departures is: {most_departures}")
```

#### What is the monthly cancellation analysis and predicted cancellations based on the minimum monthly cancellations?

```python
# Calculate the sum of cancellations for each month
cancellations_by_month = file3.groupby('MONTH')['CANCELLED'].sum().reset_index()

# Find the month with the fewest cancellations
month_with_fewest_cancellations = cancellations_by_month.loc[cancellations_by_month['CANCELLED'].idxmin()]

# Calculate the number of cancellations for each month based on the formula
cancellations_by_month['PREDICTED_CANCELLATIONS'] = cancellations_by_month['CANCELLED'] / month_with_fewest_cancellations['CANCELLED']
print(cancellations_by_month)
```

#### STATE with highest departure delay 

```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'ORIGIN_STATE': ['CA', 'TX', 'NY', 'CA', 'TX', 'NY', 'FL', 'IL', 'NV', 'AZ'],
    'DEPARTURE_DELAY': [30, 40, 10, 25, 35, 5, 15, 20, 10, 5]
}

file3 = pd.DataFrame(data)

# Group and calculate the average departure delay for each state
state_delays = file3.groupby('ORIGIN_STATE')['DEPARTURE_DELAY'].mean().reset_index()

# Sort by departure delay in descending order and get the top 10 states
top_10_states = state_delays.nlargest(10, 'DEPARTURE_DELAY')

# Create a color palette for the bar chart
colors = ['lightblue']

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_10_states['ORIGIN_STATE'], top_10_states['DEPARTURE_DELAY'], color=colors)
plt.xlabel('State')
plt.ylabel('Average Departure Delay (minutes)')
plt.title('Top 10 States with Highest Departure Delays')
plt.xticks(rotation=45)
plt.show()

```
